##### Q1 Answer via Code

```python
from qiskit.quantum_info import Operator, Pauli, SparsePauliOp
from qiskit.exceptions import QiskitError
import numpy as np
```

```python
# Q1 

op1 = SparsePauliOp.from_sparse_list(
    [("ZX", [101, 104], 1.0), ("YY", [0, 3], -1 + 1j)], num_qubits=105)
print("op1:", op1)

op2 = SparsePauliOp.from_sparse_list(
    [("XXZ", [0, 1, 4], 1 + 2j), ("ZZ", [1, 2], -1 + 1j)], num_qubits=5)
print("op2:", op2)

try:
   op3 = Pauli("ZX") + Pauli("(-1 + 1j)YY")
   print("op3", op3)
except QiskitError as e:
   print(f"op3: a QiskitError occurred: {e}")
finally:
   op3 = None

try:
   op4 = Operator(np.random.rand(105, 105), input_dims=[2, 3], output_dims=[2, 3]) 
   print("op4", op4)
except QiskitError as e:
   print(f"op4: a QiskitError occurred: {e}")
finally:
   op4 = None

print("\n\nThe answer is:")

try:
   assert(op1.dim[0] == op1.dim[1] == 2**105)
   print("answer 1 is valid because the dimensions of the operator matrix are 2**105 by 2**105, and we said that we have 105 qubits.")
   print("\t This is a big matrix, since 2**105 equals", 2**105)
   print("\n\t Using SparsePauliOp is beneficial for dealing with large quantum systems because it") 
   print("\t can represent and manipulate operators efficiently, even with hundreds of qubits,")
   print("\t as long as the number of non-zero Pauli terms is relatively small.\n")  
except AssertionError as e:
   print("answer 1 is not valid")
    
try: 
   assert(op2.dim[0] == op2.dim[1] == 2**105)
   print("answer 2 is valid")
except AssertionError as e:
   print("answer 2 is not valid; the operator's dimensions are", op2.dim, "rather than 2**105 by 2**105")
   
if op3 is None:
   print("answer 3 is not valid; we experienced QiskitError exceptions above")
if op4 is None:
   print("answer 4 is not valid; we experienced QiskitError exceptions above")
```

          1. compute the length of qc.angles

          3. check qc.num_ancillas

##### Code Answer to Q8

The answer is 2.  The QuantumCircuit class does not have an attribute called "angles".  QuantumCircuit does have the num_ancillas attribute, but that refers to a qubit count and not a parameter count.   

```python
from qiskit.transpiler import generate_preset_pass_manager
from qiskit.circuit import Parameter

theta = Parameter("theta")  # undefined number
phi = Parameter("phi")  # undefined number 

# Create and optimize circuit once
qc = QuantumCircuit(2)
qc.rx(theta, 0)
qc.rx(phi, 1)
    
qc = generate_preset_pass_manager(
    optimization_level=3, basis_gates=["u", "cx"]
).run(qc)

print("Circuit Parameters:", qc.parameters)
print("Length of Parameters attribute:", len(qc.parameters))
print("Number of Circuit Parameters:", qc.num_parameters)
print("Number of Ancilla Qubits:", qc.num_ancillas)
print("Number of Input Vars:", qc.num_input_vars)
print("Number of Vars:", qc.num_vars)
print("Number of Angles:", qc.angles)

qc.draw("mpl")
```

- **Q9: Transpile and Optimize Circuits**

      Real quantum devices are subject to noise and gate errors, so optimizing the circuits to reduce their
      depth and gate count can significantly improve the results obtained from executing those circuits.
  
      The 'generate_preset_pass_manager' function has one required positional argument, optimization_level,
      that controls how much effort the transpiler spends on optimizing circuits.

      This argument can be an integer taking one of the values 0, 1, 2, or 3.

      Which statement is true?

        1. Higher numbers mean more optimization, and usually no additional compile time

        2. optimization_level = 0 means light optimization

        3. optimization_level = 3 means max optimization,
           will likely take more time than optimization_level = 1,
           and will always produce a better circuit result than any other choice for optimization_level,
           as long as we set approximation_degree to a value < 1

        4. optimization_level = 0 inserts SWAP gates if necessary,
           optimization_level = 1 tries to select the best qubits in the graph for layout/routing and optimizes 1 qubit gates,
           optimization_level = 2 does commutative cancellation of gates,
           optimization_level = 3 resynthesizes two qubit blocks, moves measurements around to avoid SWAPs, and
                                  removes gates before measurements that would not affect the measurements 
              

##### Answer to Q9

The answer is 4.  Higher optimization levels generate more optimized circuits at the expense of compile time.  Optimization level 0 means no optimization and is used to characterize hardware.  The default value for approximation_degree is 1, which means the transpiler makes no errors in the decomposition in order to save gates and ease execution.  Lower values of approximation_degree will result in lower output accuracy.

      OpenQASM (open quantum assembly language), a machine-independent programming interface

      4. Physical qubits can be used in gate statements.

         manipulations of angles more efficient at runtime, when the hardware executing the program does not
